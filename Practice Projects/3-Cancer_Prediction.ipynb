# **Cancer Prediction**

Dataset Information:

Target Variable (y): 
- Diagnosis (M = malignant, B = benign)

Ten features (X) are computed for each cell nucleus:

1. radius (mean of distances from center to points on the perimeter)
2. texture (standard deviation of gray-scale values)
3. perimeter
4. area
5. smoothness (local variation in radius lengths)
6.  compactness (perimeter^2 / area - 1.0)
7. concavity (severity of concave portions of the contour)
8. concave points (number of concave portions of the contour)
9. symmetry
10. fractal dimension (coastline approximation - 1)

For each characteristic three measures are given:

a. Mean

b. Standard error

c. Largest/ Worst

```python
# Step 2 : import data
cancer = pd.read_csv('https://github.com/YBIFoundation/Dataset/raw/main/Cancer.csv')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Step 2: Distribution of the Target Variable
plt.figure(figsize=(8, 6))
sns.countplot(x='diagnosis', data=cancer, palette='Set2')
plt.title('Distribution of Diagnosis (Benign vs Malignant)')
plt.xlabel('Diagnosis')
plt.ylabel('Count')
plt.show()
```

```python
# Step 3: Correlation Heatmap
plt.figure(figsize=(14, 12))

# Drop non-numeric columns before calculating the correlation matrix
numeric_data = cancer.drop(columns=['id', 'diagnosis', 'Unnamed: 32'])
corr_matrix = numeric_data.corr()

# Create the heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.2)
plt.title('Correlation Heatmap of Features')
plt.show()
```

```python
y = cancer['diagnosis']
```

```python
X = cancer.drop(['id','diagnosis','Unnamed: 32'],axis=1)
```

```python
# Step 4 : train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, train_size=0.7, random_state=2529)
```

```python
# check shape of train and test sample
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# Step 5 : select model
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=5000)
```

```python
# Step 6 : train or fit model
model.fit(X_train,y_train)
```

```python
print("Intercept:", model.intercept_)
print('Coefficient:', model.coef_)
```

```python
# Step 7 : predict model
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
# Step 8 : model accuracy
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
cm= confusion_matrix(y_test,y_pred)
cm
```

```python
accuracy_score(y_test,y_pred)
```

```python
plt.figure(figsize=(4, 3))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=['Benign', 'Malignant'],
            yticklabels=['Benign', 'Malignant'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix Heatmap')
plt.show()
```

```python
print(classification_report(y_test,y_pred))
```
